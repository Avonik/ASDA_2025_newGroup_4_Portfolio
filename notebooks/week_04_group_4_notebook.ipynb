```python
url = "https://docs.google.com/spreadsheets/d/1ecopK6oyyb4d_7-QLrCr8YlgFrCetHU7-VQfnYej7JY/export?format=xlsx"
dataset = pd.ExcelFile(url, engine='openpyxl')
```

```python
dataset = pd.concat([dataset.parse(sheet_name).assign(sheet_name = sheet_name) for sheet_name in dataset.sheet_names])
dataset.reset_index(inplace = True)

dataset.sheet_name
```

```python
dataset.head(10)
```

```python
dataset.info()
```

```python
# rename name
dataset.rename(columns={'realSum': 'Price'}, inplace=True)
dataset.head()
```

```python
# drop unnamed : 0 column in dataset
dataset = dataset.drop("Unnamed: 0", axis=1)
dataset.head()
```

```python
dataset[['city', 'weekday']] = dataset["sheet_name"].str.split('_', expand = True)

dataset.sample(10)
```

```python
# unique city
dataset["city"].unique()
```

```python
dataset["weekday"].unique()
```

```python
dataset['weekday'] = dataset['weekday'].map({'weekdays': 0, 'weekends': 1})
dataset.sample(10)
```

```python
# combining column room shared & room private
def get_room_type(row):
    if row['room_shared'] == 1:
        return 'shared'
    elif row['room_private'] == 1:
        return 'private'
    else:
        return 'entire_home'
    
dataset['room_type'] = dataset.apply(get_room_type, axis=1)
dataset["room_type"].unique()
```

```python
# removing the room_shared , room_private columns 
dataset = dataset.drop(["room_shared", "room_private"], axis=1)
dataset.sample(10)
```

```python
# Instead of dummy variables, create a column where one can know if the host of the listing has only one, two to four, or more than four listings
dataset['host_listings_category'] = dataset['multi'].apply(
    lambda x: 'one' if x == 1 else ('two_to_four' if x <= 4 else 'more_than_four')
)
dataset.head(10)
```

```python
#getting statistics on categorical/object columns
obj_cols = dataset.select_dtypes(include='object').columns
print(f"This dataset has {len(obj_cols)} object columns:", list(obj_cols))
```

```python
#detailed statistics on object data
for col in dataset.select_dtypes(include='object'):
    print(f"\nColumn: {col}")
    print("Count of non-null values:", dataset[col].count()) 
    print("Number of unique values:", dataset[col].nunique())
    print("Most frequent value:", dataset[col].value_counts().idxmax())
    print("Most frequent value (frequency):", dataset[col].value_counts().max())
    print("Least frequent value:", dataset[col].value_counts().idxmin())
    print("Least frequent value (frequency):", dataset[col].value_counts().min())
```
